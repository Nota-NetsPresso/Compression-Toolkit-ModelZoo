# 2. Fine-tuning the Compressed Model.
Structured pruning may induce the performance degradation, which requires additional fine-tuning process.  
Therefore, the fine-tuning process is necessary to compensate the performance after each compression process.  
Here, `Structured Pruning` &rarr; `Fine-tuning`

### Loading the dataset

```python
from torchvision import datasets, transforms
# data will be downloaded in data_directory.
data_directory = './data'
batchsize=256

normalize = transforms.Normalize(mean=[0.507, 0.4865, 0.4409],
                                 std=[0.2673, 0.2564, 0.2761])

train_dataset = datasets.CIFAR100(root=data_directory,
                                  train=True,
                                  transform=transforms.Compose([
                                                                  transforms.RandomCrop(32, padding=4),
                                                                  transforms.RandomHorizontalFlip(),
                                                                  transforms.ToTensor(),
                                                                  normalize]),
                                  download=True)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           batch_size=batchsize,
                                           shuffle=True,
                                           pin_memory=True,
                                           num_workers=2)

test_dataset = datasets.CIFAR100(root=data_directory,
                                 train=False,
                                 transform=transforms.Compose([
                                                                  transforms.ToTensor(),
                                                                  normalize]),
                                 download=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                          batch_size=batchsize,
                                          shuffle=False,
                                          pin_memory=True,
                                          num_workers=2)
```

### Loading the compressed model

```python
compressed_model = torch.load("MobileNetV2_compressed_model.pt").to(device)
```

### Fine-tuning the compressed model

Setting the hyperparameters 

```python
import torch.nn as nn
from torch.optim.lr_scheduler import CosineAnnealingLR

lr = 0.001
epochs = 100
criterion = nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.SGD(compressed_model.parameters(), lr=lr,
                            momentum=0.9, nesterov=True, weight_decay=5e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=200, eta_min=0, verbose=False)

print("All parameters and hyperparameters are set.")
```

Fine-tuning the compressed model

```python
import tqdm
# training code
def train(model,criterion, optimizer, scheduler,loader,epoch):
    global device
    for images, labels in tqdm.tqdm(loader):
        images = images.to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    scheduler.step()
    return loss
    
# validation code
def validate(model,loader):
    global device
    model.eval()
    correct = 0.
    total = 0.
    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        with torch.no_grad():
            pred = model(images)

        pred = torch.max(pred.data, 1)[1]
        total += labels.size(0)
        correct += (pred == labels).sum().item()

    val_acc = (correct / total)*100
    model.train()
    return val_acc

# fine-tune
print("Fine-tuning start...")
best_val = 0
for epoch in range(1,epochs+1):
    
    train(compressed_model,criterion,optimizer,scheduler,train_loader,epoch)
    
    val_acc = validate(compressed_model,test_loader)
    best_val = max(val_acc,best_val)
    
    print(f"epoch[{epoch}/{epochs}] val_acc:{val_acc}")
    
print(f"Fine-tuning is done. Best Accuracy is {best_val}.")
```

### Latency comparision

```python
import onnx
import copy
def onnxExport(model, model_name, device = torch.device('cpu')):
    model = model.to(device.type)
    dummy_input = torch.ones((1,3,32,32)).to(device.type)
    torch.onnx.export(copy.deepcopy(model),dummy_input,model_name,verbose=False)

onnxExport(compressed_model, 'MobileNetV2_compressed_model.onnx')
```

```python
import onnxruntime
import numpy as np

def inference_time(model_path,run_times):
    device = torch.device('cpu')
    img_size = 32
    so = onnxruntime.SessionOptions()
    so.execution_mode = onnxruntime.ExecutionMode.ORT_SEQUENTIAL
    so.graph_optimization_level = onnxruntime.GraphOptimizationLevel.ORT_ENABLE_EXTENDED
    exproviders = [ 'CPUExecutionProvider']
    ort_session = onnxruntime.InferenceSession(model_path, so, providers=exproviders)

    input_names = ort_session.get_inputs()[0].name
    output_names = ort_session.get_outputs()[0].name
    # warm up
    dummy_input = np.random.rand(1,3,32,32).astype(np.float32)
    for _ in range(10):
        #IOBinding
        data = onnxruntime.OrtValue.ortvalue_from_numpy(dummy_input, device.type, 0)
        io_binding = ort_session.io_binding()
        io_binding.bind_input(input_names, device.type, 0, np.float32, [1, 3, img_size, img_size], data.data_ptr())
        io_binding.bind_output(output_names, device.type)

        torch.cuda.synchronize()
        with torch.no_grad():
            ort_session.run_with_iobinding(io_binding)
        torch.cuda.synchronize()

    start = torch.cuda.Event(enable_timing=True)
    end = torch.cuda.Event(enable_timing=True)

    latency = []
    for _ in range(run_times):
        dummy_input = np.random.rand(1,3,32,32).astype(np.float32)
        #IOBinding
        data = onnxruntime.OrtValue.ortvalue_from_numpy(dummy_input, device.type, 0)
        io_binding.bind_input(input_names, device.type, 0, np.float32, [1, 3, img_size, img_size], data.data_ptr())
        io_binding.bind_output(output_names, device.type)

        torch.cuda.synchronize()
        start.record() 
        with torch.no_grad():
            ort_session.run_with_iobinding(io_binding)
        end.record()
        torch.cuda.synchronize()
        time_elapsed = start.elapsed_time(end)
        latency.append(time_elapsed)
    return sum(latency)/len(latency)


original_latency = inference_time('MobileNetV2.onnx',run_times=1000)
compressed_latency = inference_time('MobileNetV2_compressed_model.onnx',run_times=1000)

print("Original Model ==> Compressed Model")
print(f"Latency: {round(original_latency,3)}ms ==> {round(compressed_latency,3)}ms (x{round(original_latency/compressed_latency,3)})")
```

### Int8 inference
Quantization is one of the model compression methods that approximates a model that uses floating point numbers by a model of low-bit width numbers.

We quantize the compressed model (FP32 &rarr; INT8) and visualize that NetsPresso Model Compressor works well with Quantization.

```python
# Define the Data Reader
from onnxruntime.quantization import quantize_dynamic, quantize_static, QuantType, QuantFormat
from onnxruntime.quantization import CalibrationDataReader
import tqdm
quant_train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           batch_size=1,
                                           shuffle=True,
                                           pin_memory=True,
                                           num_workers=2)

class CIFAR100_DataReader(CalibrationDataReader):
    def __init__(self, train_loader, model_path):
        self.enum_data = None
        self.train_loader = train_loader
        
        session = onnxruntime.InferenceSession(model_path, None)
        (_, _, height, width) = session.get_inputs()[0].shape
        self.input_name = session.get_inputs()[0].name
        self.datasize = len(train_loader)
        
        
        unconcatenated_batch_data = []
        for img, _ in train_loader:
            if len(unconcatenated_batch_data) == 500:
                break
            unconcatenated_batch_data.append(img.numpy())

        self.batch_data = np.concatenate(np.expand_dims(unconcatenated_batch_data, axis=0), axis=0)

    def get_next(self):
        if self.enum_data is None:
            self.enum_data = iter([{self.input_name: nhwc_data} for nhwc_data in self.batch_data])
        
        return next(self.enum_data, None)
    
    def rewind(self):
        self.enum_data = None
```

```python
# Quantize the original model
import copy

original_model_inference_path = "MobileNetV2_inference.onnx"
original_model_inference_quantized_path = 'MobileNetV2_inference_INT8.onnx'

dummy_input = torch.ones((1,3,32,32)).to(device)
torch.onnx.export(copy.deepcopy(model),dummy_input,original_model_inference_path,verbose=False,opset_version=14)

original_model_quantized = quantize_static(original_model_inference_path,
                                    original_model_inference_quantized_path,
                                    CIFAR100_DataReader(train_loader=quant_train_loader, model_path=original_model_inference_path),
                                    quant_format=QuantFormat.QDQ,
                                    per_channel=False,
                                    weight_type=QuantType.QInt8,
                                    optimize_model=True
                                    )
```

```python
# Quantize the compressed model
compressed_model_path = 'MobileNetV2_compressed_model.onnx'
compressed_model_quantized_path = 'MobileNetV2_compressed_model_INT8.onnx'

compressed_model_quantized = quantize_static(compressed_model_path,
                                    compressed_model_quantized_path,
                                    CIFAR100_DataReader(train_loader=quant_train_loader, model_path=compressed_model_path),
                                    quant_format=QuantFormat.QDQ,
                                    per_channel=False,
                                    weight_type=QuantType.QInt8,
                                    optimize_model=True
                                    )
```

```python
# Latency Comparision
original_model_quantized_latency = inference_time(original_model_inference_quantized_path, run_times=1000)
compressed_model_quantized_latency = inference_time(compressed_model_quantized_path,run_times=1000)
```

```python
print("Latency Comparision")
print(f"{'Original Model:':<40} {round(original_latency,3):.3f}ms")
print(f"{'Original Model (+Quantization):':<40} {round(original_model_quantized_latency,3):.3f}ms (x{round(original_latency/original_model_quantized_latency,3)})")
print(f"{'Compressed Model:':<40} {round(compressed_latency,3):.3f}ms (x{round(original_latency/compressed_latency,3):.3f})")
print(f"{'Compressed Model (+Quantization):':<40} {round(compressed_model_quantized_latency,3):.3f}ms (x{round(original_latency/compressed_model_quantized_latency,3):.3f})")
```

```python
# Accuracy comparision
import tqdm
import numpy as np

quant_test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                          batch_size=1,
                                          shuffle=False,
                                          pin_memory=True,
                                          num_workers=2)

def validate_onnx(model_path,loader):
    session = onnxruntime.InferenceSession(model_path)
    input_name = session.get_inputs()[0].name
    input_data = np.zeros((1, 3, 32, 32), np.float32)
    correct = 0
    _ = session.run([],{input_name: input_data})
    for img, target in tqdm.tqdm(loader):
        output = session.run([], {input_name: img.numpy()})
        # print(target, output)
        # print(np.argmax(output),target)
        if np.argmax(output) == target:
            correct +=1
    return correct/len(loader)
```

```python
original_model_qauntized_accuracy = validate_onnx(original_model_inference_quantized_path, quant_test_loader)*100
compressed_model_quantized_accuacy = validate_onnx(compressed_model_quantized_path, quant_test_loader)*100
```

```python
print("Accuracy Comparision")
print(f"{'Original Model:':<40} {original_model_accuracy:.2f}%")
print(f"{'Original Model (+Quantization):':<40} {original_model_qauntized_accuracy:.2f}% (-{original_model_accuracy-original_model_qauntized_accuracy:.2f})")
print(f"{'Compressed Model:':<40} {compressed_model_accuracy:.2f}% (-{original_model_accuracy-compressed_model_accuracy:.2f})")
print(f"{'Compressed Model (+Quantization):':<40} {compressed_model_quantized_accuacy:.2f}% (-{original_model_accuracy-compressed_model_quantized_accuacy:.2f})")
```
